```python
import pandas as pd
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.20, random_state=42, shuffle=True
)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 1. Instantiate (Building the frame)
model = KNeighborsClassifier(n_neighbors=5)

# 2. Fit (Memorizing the map)
model.fit(X_train, y_train)

# 3. Predict (Applying logic)
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("--- Diagnostic Confusion Matrix ---")
print(confusion_matrix(y_test, predictions))

print("\n--- Deep Metrics Validation (F1 Score, Precision, Recall) ---")
print(classification_report(y_test, predictions, target_names=iris.target_names))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
df['Species'] = [iris.target_names[i] for i in iris.target]
```

```python
def plot_pairplot(dataframe):
    plt.figure(figsize=(10, 8))
    # 'hue' colors the data points by their species class
    sns.pairplot(dataframe, hue='Species', palette='Set2', diag_kind='kde')
    plt.suptitle("Iris Feature Pairplot Matrix", y=1.02)
    plt.show()

# Call the function to test
plot_pairplot(df)
```
